```python
# load dara from csv file

import pandas as pd
import numpy as np


def get_data(No):
    
    df         = pd.read_csv('Group2.csv')
    df_temp    = df['No'+str(No)].str.split(' ',expand=True)[[0,1,2,3]]
    df_temp[2] = df_temp[2] +' ' + df_temp[3]

    df_temp.drop(3,axis=1,inplace=True)
    df_temp = df_temp.reindex(columns=[1,2,0])

    df_temp.columns = ['Student_ID'+str(No),'Name'+str(No),'Index'+str(No)]

    return df_temp



df_all = pd.concat([get_data(1),get_data(2)],axis=1)
df_all['Team'] = range(1,df_all.shape[0]+1)
df_all = df_all.reindex(columns=['Team','Student_ID1','Name1','Index1','Student_ID2','Name2','Index2'])


# shuffle the rows using the sample method
shuffled_df          = df_all.sample(frac=1, random_state=3).reset_index(drop=True)
shuffled_df['Team'] = range(1,df_all.shape[0]+1)
df_all              = shuffled_df  



df_all.to_csv('group_mapping.csv', index=True)

df_all
```

```python
import pandas as pd
from datetime import datetime, timedelta

def generate_day_timetable(day):
    start_time = datetime.strptime("14:00", "%H:%M")
    pitch_duration = timedelta(minutes=15)
    qna_duration = timedelta(minutes=5)
    break_duration = timedelta(minutes=5)

    timetable = []
    for i in range(7):
        pitch_end = start_time + pitch_duration
        qna_start = pitch_end
        qna_end = qna_start + qna_duration
        break_start = qna_end
        break_end = break_start + break_duration

        timetable.append((f'Team {i+1}', start_time.strftime("%H:%M"), pitch_end.strftime("%H:%M"), qna_start.strftime("%H:%M"), qna_end.strftime("%H:%M"), break_start.strftime("%H:%M"), break_end.strftime("%H:%M")))
        start_time = break_end

    df = pd.DataFrame(timetable, columns=['Team', 'Pitch Start', 'Pitch End', 'Q&A Start', 'Q&A End', 'Break Start', 'Break End'])
    df['Day'] = f'Day {day}'
    return df

days_data = []

for day in range(1, 7):
    day_df = generate_day_timetable(day)
    days_data.append(day_df)

timetable_df = pd.concat(days_data, axis=0, ignore_index=True)
timetable_df['Team'] = range(1,df_all.shape[0]+1)
timetable_df['Day'].replace('Day 1','18 April 2023', inplace=True)
timetable_df['Day'].replace('Day 2','19 April 2023', inplace=True)
timetable_df['Day'].replace('Day 3','25 April 2023', inplace=True)
timetable_df['Day'].replace('Day 4','26 April 2023', inplace=True)
timetable_df['Day'].replace('Day 5','2 May 2023', inplace=True)
timetable_df['Day'].replace('Day 6','3 May 2023', inplace=True)

timetable_df.to_csv('timetable.csv', index=False)

timetable_df
```
